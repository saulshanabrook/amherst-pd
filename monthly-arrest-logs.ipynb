```python

def records():
    year = None
    for (_, row) in pd.read_excel('gender-race-arrests.xlsx', header=3)[2:].iterrows():
        for (i, cell) in row.iteritems():
            if i == 'YEAR':
                year = int(cell)
                continue
            for (gender, count) in zip(['male', 'female'], cell.split(' -- ')):
                yield {
                    "year": year,
                    "race": i.lower(),
                    "gender": gender,
                    "arrests": int(count)
                }
d = pd.DataFrame(records())
d.head()
```
